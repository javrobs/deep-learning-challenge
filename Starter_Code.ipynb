## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=["EIN","NAME"],inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_vcounts=application_df["APPLICATION_TYPE"].value_counts()
app_vcounts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff=500
application_types_to_replace={}
for app in app_vcounts.keys():
  if app_vcounts[app] < cutoff:
    application_types_to_replace[app]="Other"
  else:
    application_types_to_replace[app]=app

# Replace in dataframe
application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].map(application_types_to_replace)

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_vcounts=application_df["CLASSIFICATION"].value_counts()
class_vcounts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_vcounts[class_vcounts>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff=1000
classifications_to_replace={}
for c in class_vcounts.keys():
  if class_vcounts[c] < cutoff:
    classifications_to_replace[c]="Other"
  else:
    classifications_to_replace[c]=c

# Replace in dataframe
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].map(classifications_to_replace)

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies=pd.get_dummies(application_df[["APPLICATION_TYPE","AFFILIATION","CLASSIFICATION","USE_CASE","ORGANIZATION","INCOME_AMT"]])
dummies.head()
```

```python
application_prep_df=application_df.drop(columns=["APPLICATION_TYPE","AFFILIATION","CLASSIFICATION","USE_CASE","ORGANIZATION","INCOME_AMT"])
application_prep_df["SPECIAL_CONSIDERATIONS"]=application_prep_df["SPECIAL_CONSIDERATIONS"].map({"N":0,"Y":1})
application_prep_df.head()
for c in dummies.columns:
  application_prep_df[c]=dummies[c]
application_prep_df
```

```python
# Split our preprocessed data into our features and target arrays
y=application_prep_df["IS_SUCCESSFUL"]
X=application_prep_df.drop(columns=["IS_SUCCESSFUL"])
# Split the preprocessed data into a training and testing dataset
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y)
```

```python
X_train
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()


# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=84, activation="relu", input_dim=42))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=21, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## The first model has an accuracy of 72.13%, we'll try improving this with different methods.

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
